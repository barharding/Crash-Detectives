```python
#parse out the accident # from the EventUniqueID field and create a new column.  This is to link traffic file to the KSI file
df_TC['AccNum']=df_TC['EventUniqueId'].apply(lambda x : x[-7:])
print(df_TC['AccNum'])
```

```python
# drop leading zero in the new AccNum column in df_TC to match ACCNUM column in df_KSI

df_TC['AccNum'] = df_TC['AccNum'].apply(lambda x: x.lstrip('0'))
print(df_TC['AccNum'])
```

```python
#parse out geo-coordinates in all three files
import re

df_TC["geometry2"] = df_TC["geometry"].str.extract(r"\[(.*?)\]")
df_KSI["geometry2"] = df_TC["geometry"].str.extract(r"\[(.*?)\]")
df_Redlight["geometry2"] = df_TC["geometry"].str.extract(r"\[(.*?)\]")
```

```python
# split geometry2 into two columns: long & Lat

df_TC[['Longitude', 'Latitude']] = df_TC['geometry2'].str.split(',', 1, expand=True)
df_KSI[['Longitude', 'Latitude']] = df_KSI['geometry2'].str.split(',', 1, expand=True)
df_Redlight[['Longitude', 'Latitude']] = df_Redlight['geometry2'].str.split(',', 1, expand=True)

print(df_Redlight['Longitude'], df_Redlight['Latitude'])
```
