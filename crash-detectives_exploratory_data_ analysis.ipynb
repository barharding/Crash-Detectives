```python
#parse out the accident # from the EventUniqueID field and create a new column.  This is to link traffic file to the KSI file
df_TC['AccNum']=df_TC['EventUniqueId'].apply(lambda x : x[-7:])
print(df_TC['AccNum'])
```

```python
# drop leading zero in the new AccNum column in df_TC to match ACCNUM column in df_KSI

df_TC['AccNum'] = df_TC['AccNum'].apply(lambda x: x.lstrip('0'))
print(df_TC['AccNum'])
```
